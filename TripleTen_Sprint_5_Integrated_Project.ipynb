## Step 5: Testing the Hypotheses

### Hypothesis 1: The average user ratings for the xOne and PC platforms are the same

H0 (null) - The average user rating of the Xbox One and PC platforms is the same.

H1 (alternative) The average user rating of the Xbox One and PC platforms is different.

Since we are testing the hypothesis that the user ratings of the Xbox One and PC platforms are either the same or different, we use a two-tailed test. The chosen significance level (alpha) is 0.05. This means that there is a 5% chance of rejecting the null hypothesis when it is true. This significance level because is a commonly used value that provides a good balance between the risk of making a Type I error and the power of the test (the probability of rejecting the null hypothesis when it is false). If the p-value is greater than or equal to 0.05, we will fail to reject the null hypothesis and conclude that there is a difference between the tested values.

```python
# Create a dataset with data from the top 5 platforms from the last 5 years
top = df.query('platform in ["PS4", "XOne", "3DS", "PC", "WiiU"] and year_of_release >= 2012')

# Check for missing values in the data
top.isnull().sum()
```

```python
# Drop missng values from the data
top = top.dropna()

# Check that missing values have been dropped
top.isnull().sum()
```

```python
# Create variables for the xone and pc platforms
xone = top[top['platform'] == 'XOne']
pc = top[top['platform'] == 'PC']

# Calculate mean values for each platform's user scores
xone_user_score_mean = xone['user_score'].mean()
pc_user_score_mean = pc['user_score'].mean()

# Print mean values for each platform's user scores
print('XOne User Score Mean:', xone_user_score_mean)
print('PC User Score Mean:', pc_user_score_mean)

# Test hypothesis that the average user score for the xOne and PC platforms are equal
alpha = 0.05
results = ttest(xone['user_score'], pc['user_score'], equal_var=False)
print('p-value:', results.pvalue)
if results.pvalue < alpha:
    print('We reject the null hypothesis')
else:
    print('We fail to reject the null hypothesis')
```

### Hypothesis 2: The average rating for the action and sports genres is different
